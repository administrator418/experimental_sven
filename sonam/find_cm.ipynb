```python
from PIL import Image
import numpy as np
```

```python
def find_cm(image_path):
    img = Image.open(image_path)
    img_gray = img.convert('L')
    img_array = np.array(img_gray)
    threshold = 128
    binary_image = np.where(img_array > threshold, 1, 0)
    x_indices, y_indices = np.nonzero(binary_image)
    center_of_mass_x = np.mean(x_indices)
    center_of_mass_y = np.mean(y_indices)
    return(center_of_mass_x, center_of_mass_y)
```

```python
image_path = 'reference_defocus_Vortex.bmp'
```

```python
center_x, center_y = find_cm(image_path)
print(center_x, center_y)
```

```python
import matplotlib.pyplot as plt
img = Image.open(image_path)
plt.imshow(img, cmap='gray')
plt.scatter(center_y, center_x, color = 'red', marker='x', s=100)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Read the BPM image file
image = plt.imread('reference_defocus_Vortex.bmp')

# Threshold the image to extract the beam profile
threshold = 0.5  # Adjust the threshold value as needed
binary_image = np.where(image > threshold, 128, 0)

# Calculate the centroid of the beam profile
x_indices, y_indices = np.nonzero(binary_image)
centroid_x = np.mean(x_indices)
centroid_y = np.mean(y_indices)

print("Centroid (x, y):", centroid_x, centroid_y)
```
